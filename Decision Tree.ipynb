# Iris Dataset

```python
#importing necessary lib's
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.data
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
d=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
d.head()
```

```python
d['target']=iris.target
```

```python
d.head()
```

```python
d0=d[:50]
```

```python
d1=d[50:100]
```

```python
plt.xlabel("sepal length")
plt.ylabel("Sepal width")
plt.scatter(d0['sepal length (cm)'],d0['sepal width (cm)'],marker='*',color='blue')
plt.scatter(d1['sepal length (cm)'],d1['sepal width (cm)'],marker='.',color='green')
```

```python
x=d.drop(['target'],axis=1)
y=d['target']
```

```python
#split the data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#modeling algorithm
from sklearn.tree import DecisionTreeClassifier
```

```python
dtr=DecisionTreeClassifier()
```

```python
dtr.fit(x_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,8))
tree.plot_tree(dtr,filled=True)
```

```python
dtr.predict([[5.1,3.5,2.4,0.3]])
```

```python
dtr.score(x_train,y_train)
```

```python
dtr.score(x_test,y_test)
```

```python
y_pred=dtr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
AC=accuracy_score(y_test,y_pred)
AC
```

# Titanic Dataset

```python
d=pd.read_csv("titanic_train.csv")
```

```python
d.columns
```

```python
d.drop(['Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked','SibSp', 'Name','PassengerId'],axis=1,inplace=True)
```

```python
d.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
d['Sex']=le.fit_transform(d['Sex'])
```

```python
d.head()
```

```python
d=d.drop(['sex'],axis=1)
```

```python
d.isnull().sum()
```

```python
d['Age']=d['Age'].fillna(d['Age'].median())
```

```python
d.isnull().sum()
```

```python
sns.countplot(x='Survived',hue='Pclass',data=d,palette='dark:red')
```

```python
X=d.drop(['Survived'],axis=1)
Y=d['Survived']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy')
```

```python
dtc.fit(x_train,y_train)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(dtc,filled=True)
```

```python
dtc.predict([[3,0,34.0]])
```

```python
dtc.score(x_train,y_train)
```

```python
dtc.score(x_test,y_test)
```

```python
y_pred=dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
AC=accuracy_score(y_test,y_pred)*100
print(f"Accuracy score :{AC:.2f}")
```
